```python
def extract_text(text):
    
    # Create reader variable that will read the pdffileobj
    pdf_reader = PyPDF2.PdfReader(text)
    
    # This will store the number of pages of this pdf file
    x = len(pdf_reader.pages)
    
    # Create a variable that will select the selected number of pages
    page = pdf_reader.pages[0]
    text = page.extract_text()
    
    complete_text = ''
    for i in range(0, x):
        page = pdf_reader.pages[i]
        text = page.extract_text()
        complete_text = complete_text + text 
    
    # converting enitre text to lowercase
    raw_doc = complete_text.lower() 
    return raw_doc
```

```python
# Tokenization
def LemTokens(tokens):
    return [lemmer.lemmatize(token) for token in tokens]
```

```python
our_no_string = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~  \n \n  \n \n \n  \n\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_{|}~\t\n�u '
remove_punc_dict = dict((ord(punct), None) for punct in string.punctuation + '\n')
```

```python
# Remoce stop words
def LemNormalize(text):
    return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punc_dict)))
```

```python
greet_inputs = ('hej', 'hejsan', 'hur är läget?', 'halloj?', 'Hej', 'Hejsan', 'Hur är läget?', 'Halloj')
```

```python
greet_responses = ('Hej!', 'Hejsan!', 'Hej där!', 'Hej på dig!!')
```

```python
def greet(sentence):
    for word in sentence.split():
        if word.lower() in greet_inputs:
            return random.choice(greet_responses)
```
